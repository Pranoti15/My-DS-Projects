```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded dataset
iris_data = pd.read_csv('/content/Iris.csv')

# Display the first few rows to understand the structure of the dataset
iris_data.head()
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Question 1: Distribution of sepal lengths and widths among the Iris species
plt.figure(figsize=(10, 4))

# Sepal Length distribution
plt.subplot(1, 2, 1)
sns.boxplot(data=iris_data, x='Species', y='SepalLengthCm', hue='Species', palette='Set2', dodge=False)
plt.title('Distribution of Sepal Length among Iris Species')
plt.xlabel('Iris Species')
plt.ylabel('Sepal Length (cm)')

# Sepal Width distribution
plt.subplot(1, 2, 2)
sns.boxplot(data=iris_data, x='Species', y='SepalWidthCm', hue='Species', palette='Set3', dodge=False)
plt.title('Distribution of Sepal Width among Iris Species')
plt.xlabel('Iris Species')
plt.ylabel('Sepal Width (cm)')

plt.tight_layout()
plt.show()
```

**Insights:**

---


**Sepal Length:**
Iris-virginica generally has the longest sepals.
Iris-setosa has shorter sepals compared to other species.
There is some overlap between Iris-versicolor and Iris-virginica.

---



**Sepal Width:**
Iris-setosa typically has wider sepals compared to other species.
Iris-versicolor and Iris-virginica show more overlap in their sepal width distributions, with narrower widths.

```python
# Question 2: Variation of petal length across different Iris species
plt.figure(figsize=(6, 4))
sns.boxplot(data=iris_data, x='Species', y='PetalLengthCm', hue='Species', palette='coolwarm')
plt.title('Variation of Petal Length across Iris Species')
plt.xlabel('Iris Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

**Insights**:

---


Iris-setosa has significantly shorter petals compared to the other two species.

---


Iris-versicolor and Iris-virginica show distinct petal lengths, with Iris-virginica having longer petals.

---


Petal length serves as a distinguishing feature between species, particularly between Iris-setosa and the others.

```python
# Question 3: Relationship between sepal length and petal length
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris_data, x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='viridis', s=50)
plt.title('Relationship between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

**Insights**:

---


There is a positive correlation between sepal length and petal length.

---


Iris-setosa forms a distinct cluster with smaller sepal and petal lengths.

---


Iris-versicolor and Iris-virginica show overlap but can be distinguished, with Iris-virginica having generally larger measurements.

```python
# Question 4: Variation of petal width across different Iris species
plt.figure(figsize=(6, 4))
sns.boxplot(data=iris_data, x='Species', y='PetalWidthCm', hue='Species', palette='cool', dodge=False)
plt.title('Variation of Petal Width across Iris Species')
plt.xlabel('Iris Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

insights:

---


Iris-setosa has the smallest petal widths, with minimal overlap with other species.

---


Iris-versicolor has intermediate petal widths with some variability.

---


Iris-virginica has the largest petal widths, showing a clear distinction from Iris-setosa.

```python
# Question 5: Pairwise relationships between features
# Check for data types and null values in the dataset
iris_data.info()

# Check for any non-numeric or unexpected values in the numerical columns
iris_data.describe(include='all')
```

```python
# Pairwise scatterplot with histograms for better analysis
sns.pairplot(iris_data.drop(columns=['Id']), hue='Species', palette='husl', diag_kind='hist', height=2.5)
plt.suptitle('Pairwise Relationships Between Features', y=1.02)
plt.show()
```

The updated pairplot shows pairwise relationships between features and distributions of individual features:


---


**Distinct Clusters**: Iris-setosa is well-separated from Iris-versicolor and Iris-virginica in most feature combinations.

---


**Overlap**: Iris-versicolor and Iris-virginica show overlap in some feature pairs but can still be distinguished.

---


**Patterns**: Features like petal length and petal width are highly correlated and contribute to species differentiation.
