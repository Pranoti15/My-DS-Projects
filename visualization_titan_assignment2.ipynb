```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
titanic_data = pd.read_csv('/content/train.csv')

# Display the first few rows of the dataset to understand its structure
titanic_data.head()
```

```python
# 1. Distribution of passenger ages on the Titanic
plt.figure(figsize=(8, 4))
sns.histplot(titanic_data['Age'], bins=40, kde=True, color='blue')
plt.title('Distribution of Passenger Ages on the Titanic', fontsize=14)
plt.xlabel('Age', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insights**: The distribution of passenger ages shows that the majority of passengers were adults i.e., there is a peak around 20 to 30 years, indicating many passengers in that age group.

```python
# 2. Survival rate based on passenger class (Pclass)
plt.figure(figsize=(6, 3))
sns.barplot(x='Pclass', y='Survived', hue='Pclass', data=titanic_data, palette='Set2', dodge=False)
plt.title('Survival Rate by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class (Pclass)', fontsize=10)
plt.ylabel('Survival Rate', fontsize=10)
plt.ylim(0, 1)
plt.show()
```

**Insights**: Passengers in 1st class had the highest survival rate, followed by second-class passengers, with third-class passengers having the lowest survival rate. This highlights the influence of socio-economic status on survival during the Titanic disaster.

```python
# 3. Relationship between fare paid and likelihood of survival
plt.figure(figsize=(6, 3))
sns.boxplot(y='Fare', x='Survived', hue='Survived', data=titanic_data, palette='Set2', dodge=False)
plt.title('Fare Paid vs. Likelihood of Survival', fontsize=14)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=10)
plt.ylabel('Fare Paid', fontsize=10)
plt.yscale('log')  # Use log scale for better visualization of fare distribution
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insights** : The boxplot indicates that passengers who paid higher fares had a higher likelihood of survival. The distribution shows a clear skew towards higher fares for survivors, suggesting that wealthier passengers had better access to lifeboats or assistance during the disaster.

```python
# 4. Survival rate difference between male and female passengers
plt.figure(figsize=(6, 3))
sns.barplot(x='Sex', y='Survived', hue='Sex', data=titanic_data, palette='coolwarm', dodge=False)
plt.title('Survival Rate by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=11)
plt.ylabel('Survival Rate', fontsize=10)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insights** : The survival rate was significantly higher for female passengers compared to male passengers. This aligns with the "women and children first" policy during evacuation efforts. Female passengers were prioritized for lifeboats, resulting in a markedly higher survival rate.

```python
# 5. Impact of SibSp and Parch on survival
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# SibSp
sns.barplot(x='SibSp', y='Survived', hue='SibSp', data=titanic_data, palette='muted', ax=axes[0], dodge=False)
axes[0].set_title('Survival Rate by Number of Siblings/Spouses (SibSp)', fontsize=14)
axes[0].set_xlabel('Number of Siblings/Spouses', fontsize=12)
axes[0].set_ylabel('Survival Rate', fontsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[0].set_ylim(0, 1)

# Parch
sns.barplot(x='Parch', y='Survived', hue='Parch', data=titanic_data, palette='muted', ax=axes[1], dodge=False)
axes[1].set_title('Survival Rate by Number of Parents/Children (Parch)', fontsize=14)
axes[1].set_xlabel('Number of Parents/Children', fontsize=12)
axes[1].set_ylabel('Survival Rate', fontsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

**Insights** : **SibSp (Number of siblings/spouses)**: Passengers with 1 or 2 siblings/spouses aboard had a higher survival rate. However, having 3 or more siblings/spouses seemed to decrease the likelihood of survival, likely due to resource limitations or difficulties during evacuation.
**Parch (Number of parents/children)**: Passengers with 1 or 2 parents/children aboard showed a higher survival rate. Having no or many family members (3 or more) aboard was associated with lower survival rates.

```python
# 6. Survival rates by embarkation port
plt.figure(figsize=(6, 3))
sns.barplot(x='Embarked', y='Survived', hue='Embarked', data=titanic_data, palette='pastel', dodge=False)
plt.title('Survival Rates by Embarkation Port', fontsize=14)
plt.xlabel('Embarkation Port (C = Cherbourg, Q = Queenstown, S = Southampton)', fontsize=10)
plt.ylabel('Survival Rate', fontsize=10)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insights** :

---


**C**: Passengers who embarked from Cherbourg had the highest survival rate. This could be due to a higher proportion of first-class passengers embarking from this port.

---


**Q**: Passengers from Queenstown had a moderate survival rate.

---


**S**: Passengers from Southampton had the lowest survival rate, possibly due to a larger number of third-class passengers boarding there.
